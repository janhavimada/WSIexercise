# Merging the files

```python
#First I would merge the sample and cell metadata files together to get one metadata
metadata_merged= metadata.merge(sample_metadata,on="Donor")
```

```python
metadata_merged
```

### We can observe that the merge isn't running well, it is possible that the data types are not consistent, hence both columns need to be coerced to a string type. Also the barcode index is lost on merging, hence creating a new column with barcodes for easier manipulation

```python
metadata.Donor = metadata.Donor.astype(str)
sample_metadata.Donor = sample_metadata.Donor.astype(str)
sample_metadata.Age=sample_metadata.Age.astype(str)
metadata["Barcodes"]=metadata.index
```

```python
metadata_merged= metadata.merge(sample_metadata,on="Donor")
```

```python
metadata_merged.index=metadata_merged.Barcodes
metadata_merged=metadata_merged.drop(columns="Barcodes")
```

```python
metadata_merged
```

```python
#Next I would add this to the obs column of adata to populate it
#First I subset only the cells having metadata and create a copy of the adata
adata=adata_raw[adata_raw.obs.index.isin(metadata_merged.index)].copy()
```

```python
adata.obs= adata.obs.merge(metadata_merged, left_index=True,right_index=True)
```

```python
adata.obs
```

# The AnnData object is ready with the required metadata

# QC of the data

```python
#Check the max value of the adata to figure out if it is raw or processed
adata.X.max()
```

## Clearly the value is a decimal hence the counts are most likely processed, however the max count is still very high, it may require reprocessing, this can be checked by the QC plots and its distribution

```python
## Basic QC steps
sc.pp.filter_cells(adata, min_genes=200)##filtering cells where min genes less than N
sc.pp.filter_genes(adata, min_cells=3) ##filtering genes expressed in less than 3 cells
```

```python
adata.var['mt'] = adata.var_names.str.startswith('MT-')  # annotate the group of mitochondrial genes as 'mt'

# Calculating QC metrics
sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
```

```python
display(adata.var.head())
```

```python
sc.pl.violin(adata, ['n_genes_by_counts', 'total_counts', 'pct_counts_mt'],
             jitter=0.4, multi_panel=True)

sc.pl.scatter(adata, x='total_counts', y='pct_counts_mt')
sc.pl.scatter(adata, x='total_counts', y='n_genes_by_counts')
```

### From the plots it seems the mitochondrial genes have already been filtered, there also seem to be some outliers with really high total counts, these can be filtered

```python
adata = adata[adata.obs.n_genes_by_counts < 7500, :] 
adata = adata[adata.obs.total_counts < 40000, :]
```

```python
sc.pl.violin(adata, ['n_genes_by_counts', 'total_counts'],
             jitter=0.4, multi_panel=True)
sc.pl.scatter(adata, x='total_counts', y='n_genes_by_counts')
```

```python
#Shape of dataset after basic QC
adata.X.shape
```

```python
# # Total-count normalize (library-size correct) the data matrix X to 10,000 reads per cell, so that counts become comparable among cells.
sc.pp.normalize_total(adata, target_sum=1e4) 

# # # # #Logarithmize the data
sc.pp.log1p(adata)

# # # # #Identify highly-variable genes
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5) 
sc.pl.highly_variable_genes(adata)
```

```python
sc.pp.scale(adata, zero_center=False)
```

```python
adata.X.max() #The max count has decreased and scaled
```

# Dimensionality reduction with PCA

```python
sc.tl.pca(adata, svd_solver='arpack',use_highly_variable=True,zero_center=False)
sc.pl.pca(adata, color='Donor')
sc.pl.pca_variance_ratio(adata,)
```

```python
# From the elbow plot we can see that even 5 PCs explain most of the variance, hence that can be used further
```

```python
sc.pp.neighbors(adata, n_neighbors=50, n_pcs=5)
```

```python
# computing UMAP
sc.tl.umap(adata)
```

```python
# plotting UMAP
sc.pl.umap(adata,color='Donor')
```

```python
# computing tSNE
sc.tl.tsne(adata, n_pcs=5)
```

```python
# plotting tSNE
sc.pl.tsne(adata,color=['Donor'])
```
